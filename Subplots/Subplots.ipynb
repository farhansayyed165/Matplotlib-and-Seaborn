```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data = pd.read_csv('data_lang.csv')
ages = data['Age']
dev = data['All_Devs']
py = data['Python']
js = data['JavaScript']

```

```python
data.head()
```

```python
plt.style.available
```

```python
## Plotting 2 line plots in different windows
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

plt.style.use('seaborn')

ax1.plot(ages,dev,label = 'All Dev Salaries', linestyle = '--', color='#444444')
ax2.plot(ages,py,label='Python Salaries')
ax2.plot(ages,js,label = 'JavaSript Salaries')


ax1.legend()
ax1.set_title('Median Salaries by Ages')
ax1.set_ylabel('Median Salary(USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary(USD)')

plt.tight_layout()
plt.show()
```
